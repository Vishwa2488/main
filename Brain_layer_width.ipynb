#Usage of code

To get the width values and intersection points, call newmapper function with the parameters:

#Usage of mapper() function
1.   **link:** link of json file containing the data
2.   **basecoordinatesindex:** The index of lateral ventricle as per the order in which the features are stored in the json file. This can be got by using the 'seelateralventricle' function defined below. Give the index of the lateral ventricle from which you want the normal to be drawn.

3.   **withsubpiallayer:** The default value of this parameter is 0 meaning subpial layer data won't be visible by default. Subpial layer data can have None type data too. So to avoid it, the default value is 0. But if you give the value as 1, then you can also get the data for subpial layer with some data as None. None is assigned to data that cannot be found with the data in the json file.
4.   **motion:** The default value of this parameter is 1. This parameter states the direction in which the normals are drawn on the given lateral ventricle for finding the data. Giving this as -1 will make the index of coordinates on lateral ventricle boundary to decrease with every iteration.
5. **startingindex:** The default value of this parameter is 0. This gives the first index of list having the coordinates of lateral ventricle at which normal is drawn. This can be changed if 0 index is at a point where there is no required layer for data retrival.


#Usage of seelateralventricle function

Give the path of the json file as parameter to the function and it will show the brain layer and the lateral ventricle and a small starting part. **This starting part will help you understand where the 0 index is present in the lateral ventricle and hence helps you to approximate the startingindex parameter**.

#Computation

```python
def slope(coordinates, i):
    n=len(coordinates)

    if coordinates[i][0] == coordinates[i-1][0] and coordinates[(i+1)%n][0] == coordinates[i][0]:
        return 0
    if coordinates[i][0]==coordinates[i-1][0]:
        m2 = (coordinates[(i+1)%n][1]-coordinates[i][1])/(coordinates[(i+1)%n][0]-coordinates[i][0])
        if math.fabs(m2)<1:
            return None
        theta = math.fabs(math.atan(m2))

        return math.tan(math.pi/4 - theta/2)

    if coordinates[(i+1)%n][0]==coordinates[i][0]:
        m1 = (coordinates[i][1]-coordinates[i-1][1])/(coordinates[i][0]-coordinates[i-1][0])
        if math.fabs(m1)<1:
            return None
        theta = math.fabs(math.atan(m1))

        return math.tan(math.pi/4 - theta/2)

    m1 = (coordinates[i][1]-coordinates[i-1][1])/(coordinates[i][0]-coordinates[i-1][0])
    m2 = (coordinates[(i+1)%n][1]-coordinates[i][1])/(coordinates[(i+1)%n][0]-coordinates[i][0])


    angle = math.fabs(math.atan((m2-m1)/(1+m1*m2))*180/math.pi)
    if angle > 45:
        return None
    if m1 == m2:
        return -1/m1
    if m1 == -m2:
        return float('inf')
    sec1 = math.sqrt(m1*m1+1)
    sec2 = math.sqrt(m2*m2+1)
    final1 = (sec2*m1-sec1*m2)/(sec2-sec1)
    final2 = (sec2*m1+sec1*m2)/(sec2+sec1)

    if math.fabs((final1-m1)/(1+final1*m1))>1:
        return final1
    else:
        return final2

def get_normaleqn(coordinates,i):
    slope1=slope(coordinates,i)
    if slope1 == None:
        return None
    return {'coordinates':coordinates[i],'slope':slope1,'intercept':coordinates[i][1]-slope1*coordinates[i][0]}
def checkside(point, slope, intercept):
    return point[1]-slope*point[0]-intercept

def distance(point1,point2):
    return ((point1[0]-point2[0])**2 + (point1[1]-point2[1])**2)**0.5

def pointsandline(point1,point2,m,c):
    if point2[0] == point1[0]:
        return [point1[0], c+m*point1[0]]
    slope = (point2[1]-point1[1])/(point2[0]-point1[0])
    intercept = point2[1]-point2[0]*slope
    return [(intercept-c)/(m-slope),m*(intercept-c)/(m-slope)+c]
```

```python

def limitsfinder(features):
    minimax=[float('inf'),0,0]
    maximax=-float('inf')
    minimay=float('inf')
    maximay=[-float('inf'),0,0]
    for j in range(len(features)):
        coordinates = features[j]['geometry']['coordinates'][0]
        miny = min([a[1] for a in features[j]['geometry']['coordinates'][0]])
        maxx = max([a[0] for a in features[j]['geometry']['coordinates'][0]])


        for i in range(len(coordinates)):
            if coordinates[i][0] < minimax[0]:
                minimax = [coordinates[i][0],j,i]

            if coordinates[i][1]>maximay[0]:
                maximay = [coordinates[i][1],j,i]

        if miny < minimay:
            minimay = miny
        if maxx > maximax:
            maximax = maxx

    limits = [0,0,0,0]
    limits[0] = minimax[0] - 1000
    limits[1] = maximax + 1000
    limits[2] = minimay - 1000
    limits[3] = maximay[0] + 1000

    return limits

def layerfinder(link):
    with open(link) as f:
        return json.load(f)





```

```python

def indices_map_generator(features,withsubpiallayer = 0):
    map = {'ventricular zone':0,'subventricular zone':1,'intermediate zone':2,'subplate zone':3,'cortical plate':4,'marginal zone':5,'subpial granular zone':6}
    if withsubpiallayer ==1:
        mappingindices = [{'name':'ventricular zone'},{'name' : 'subventricular zone'},{'name' : 'intermediate zone'},{'name': 'subplate zone'},{'name': 'cortical plate'},{'name':'marginal zone'},{'name' : 'subpial granular zone'}]
        for i in range(len(features)):
            name=features[i]['properties']['data']['name']

            if name == 'ventricular zone' or name == 'subplate zone' or  name == 'intermediate zone' or name == 'marginal zone' or name == 'subpial granular zone' or name == 'subventricular zone' or name == 'cortical plate':

                if 'indices' in mappingindices[map[name]]:
                    mappingindices[map[name]]['indices'].append(i)
                else:
                    mappingindices[map[name]]['indices'] = [i]
    else:
        mappingindices = [{'name': 'ventricular zone'},{'name' : 'subventricular zone'},{'name' : 'intermediate zone'},{'name': 'subplate zone'},{'name': 'cortical plate'},{'name':'marginal zone'}]
        for i in range(len(features)):
            name=features[i]['properties']['data']['name']
            if name == 'ventricular zone' or name == 'subplate zone' or  name == 'intermediate zone' or name == 'marginal zone' or name == 'subventricular zone' or name == 'cortical plate':

                if 'indices' in mappingindices[map[name]]:
                    mappingindices[map[name]]['indices'].append(i)
                else:
                    mappingindices[map[name]]['indices'] = [i]
    return mappingindices

def finddirection(basecoordinates,i,normaleqn):
    m = normaleqn['slope']
    direction = (1/math.sqrt(1+m*m),m/math.sqrt(1+m*m))
    direction1 = (-direction[0], -direction[1])
    lst=[]
    lst1 = []
    n = len(basecoordinates)
    for j in range(len(basecoordinates)-1):
        if j==i:
            continue
        x = checkside(basecoordinates[j],normaleqn['slope'],normaleqn['intercept'])
        y = checkside(basecoordinates[(j+1)%n],normaleqn['slope'],normaleqn['intercept'])
        if x==0 and y==0:
            z = checkside(basecoordinates[(j+2)%n],normaleqn['slope'],normaleqn['intercept'])
            w = checkside(basecoordinates[(j-1)%n],normaleqn['slope'],normaleqn['intercept'])
            if z*w<0:
                continue

        if y == 0:
            continue
        if x == 0:
            z = checkside(basecoordinates[j-1],normaleqn['slope'],normaleqn['intercept'])
            if z*y<0:
                reqdpoint = basecoordinates[j]
                if (reqdpoint[0]-basecoordinates[i][0])*direction[0]+(reqdpoint[1]-basecoordinates[i][1])*direction[1]>0:
                    lst.append(reqdpoint)
                    lst.append(reqdpoint)
                else:
                    lst1.append(reqdpoint)
                    lst1.append(reqdpoint)


        if x*y<0:
            reqdpoint = pointsandline(basecoordinates[j],basecoordinates[j+1],normaleqn['slope'],normaleqn['intercept'])

            if ((reqdpoint[0]-basecoordinates[i][0])*direction[0]+(reqdpoint[1]-basecoordinates[i][1])*direction[1])>0:
                lst.append(reqdpoint)
            else:
                lst1.append(reqdpoint)

    if len(lst1)%2 == 0:
        return direction1
    else:
        return direction









def finder(basecoordinates,i,normaleqn,coordinates,direction):
    lst = []

    n= len(coordinates)
    for j in range(len(coordinates)):
        x = checkside(coordinates[j],normaleqn['slope'],normaleqn['intercept'])
        y = checkside(coordinates[(j+1)%n],normaleqn['slope'],normaleqn['intercept'])
        if x*y<0:
            reqdpoint = pointsandline(coordinates[j],coordinates[(j+1)%n],normaleqn['slope'],normaleqn['intercept'])

            if (direction[0]*(reqdpoint[0]-basecoordinates[i][0])+direction[1]*(reqdpoint[1]-basecoordinates[i][1]))>0:
                lst.append([reqdpoint,distance(reqdpoint,basecoordinates[i])])

    lst.sort(key = lambda lst: lst[1])



    if len(lst) ==0:
        return None

    if lst[0][1]>100 and len(lst)==1:

        return lst[0]
    if len(lst)==1:
        return None
    return lst[1]


def getnext_i(coordinates, i,direction = 1, DISTANCE=500):

    if direction>0:
        n=len(coordinates)

        diff=0
        Distance = 0
        while diff<=n-1:


            Distance += distance(coordinates[(i+diff)%n],coordinates[(i+diff+1)%n])

            if Distance > DISTANCE:
                return (i+diff+1)%n
            diff+=1
    else:
        n=len(coordinates)

        diff=0
        Distance=0
        while diff<=n-1:


            Distance += distance(coordinates[i-diff],coordinates[i-diff-1])
            if Distance > DISTANCE:
                if i-diff-1<0:
                    return n+i-diff-1
                return i-diff-1
            diff+=1
    return None




def mapper(link,basecoordinatesindex,withsubpiallayer=0,motion=1,startingindex = 0):

    with open('/content/B_141_FB40-SL_283-ST_NISL-SE_847_.json') as file1:
        layer = json.load(file1)
    stopper =1
    features = layer['features']
    limits = limitsfinder(features)
    basecoordinates = features[basecoordinatesindex]['geometry']['coordinates'][0]
    i = startingindex
    N = len(basecoordinates)
    mappingindices = indices_map_generator(features,withsubpiallayer)

    map1 = {'ventricular zone':0,'subventricular zone':1,'intermediate zone':2,'subplate zone':3,'cortical plate':4,'marginal zone':5,'subpial granular zone':6}
    ORDER = [ 'subpial granular zone','marginal zone','cortical plate'
                    ,'subplate zone', 'intermediate zone', 'subventricular zone', 'ventricular zone']
    flag = 1
    LENGTH = 0
    FINALPOINTS = []
    WIDTHDATA = []
    x=0
    count =0
    while x<=N and stopper == 1:

        flag =1
        finalpoints = {}
        distancelist = []
        finalwidths = {}
        normaleqn = get_normaleqn(basecoordinates,i)
        if normaleqn == None:
            i1 = i
            i = getnext_i(basecoordinates,i,motion)

            if i<i1:
                x+=(N-i)+i1
            else:
                x+= i-i1
            continue
        direction = finddirection(basecoordinates,i,normaleqn)

        for indicesdata in mappingindices:

            name = indicesdata['name']
            lst1 = []
            for j in indicesdata['indices']:
                coordinates = features[j]['geometry']['coordinates'][0]
                pointanddistance = finder(basecoordinates,i,normaleqn,coordinates,direction)

                if pointanddistance != None:
                    lst1.append(pointanddistance)

            lst1.sort(key = lambda lst1: lst1[1] )

            if len(lst1) == 0 and name =='ventricular zone':
                count+=1
                if count>10:
                    stopper =0

            if withsubpiallayer == 1 and map1[name] == 6:
                if len(lst1) == 0:
                    finalpoints[name] = None
                    finalwidths[name] = None
                    continue
                else:
                    reqdpoint = lst1[0]
                    finalpoints[name] = reqdpoint[0]
                    distancelist.append(reqdpoint[1])

                    if distancelist[-1]<distancelist[-2]:
                        flag = 0
                        break
                    dist = distancelist[-1] - distancelist[-2]
                    if dist >2000:
                        finalwidths[name] = None
                        finalpoints[name] = None
                        continue
                    finalwidths[name] = dist

                    continue


            if len(lst1) == 0 and map1[name]<= 5:
                flag= 0
                break
            else:


                reqdpoint = lst1[0]

                if name == 'ventricular zone':
                    count = 0
                    finalwidths[name] = distance(basecoordinates[i],reqdpoint[0])
                    distancelist.append(finalwidths[name])
                    continue

                finalpoints[name] = reqdpoint[0]
                distancelist.append(reqdpoint[1])

                if distancelist[-1]<distancelist[-2]:
                    flag=0
                    break
                finalwidths[name] = distancelist[-1] - distancelist[-2]

        if flag ==0:
            i1 = i
            i = getnext_i(basecoordinates,i,motion)

            if i<i1:
                x+=(N-i)+i1
            else:
                x+= i-i1
            continue

        FINALPOINTS.append({'basecoordinates':basecoordinates[i],'points':finalpoints})
        WIDTHDATA.append({'basecoordinates':basecoordinates[i],'widths':finalwidths})

        i1 = i
        i = getnext_i(basecoordinates,i,motion)

        if i<i1:
            x+=(N-i)+i1
        else:
            x+= i-i1

    fig = go.Figure()

    for polygon in features:
        coordinates = polygon['geometry']['coordinates'][0]
        verts = coordinates + [coordinates[0]]
        x, y = zip(*verts)


        fig.add_trace(go.Scatter(
            x=x,
            y=y,
            mode='lines',
            fill='toself',
            fillcolor='#' + polygon['properties']['data']['color_hex_triplet'],
            name = polygon['properties']['data']['name'],
            line=dict(color='#000000', width=2),
            showlegend = True
        ))
    lstx = []
    lsty = []
    for data in FINALPOINTS:
        lstx.append(data['basecoordinates'][0])
        lsty.append(data['basecoordinates'][1])
        for name in ORDER:
            if name in data['points']:
                point = data['points'][name]
                if point == None:
                    continue
                fig.add_trace(go.Scatter(
                                    x=[point[0],data['basecoordinates'][0]],
                                    y=[point[1],data['basecoordinates'][1]],
                                    mode='lines',
                                    line=dict(color='#000000', width=2),
                                    showlegend = True
                                ))
                break

        for name in data['points']:
            if data['points'][name] == None:
                continue
            lstx.append(data['points'][name][0])
            lsty.append(data['points'][name][1])



    fig.add_trace(go.Scatter(x=lstx, y=lsty,mode='markers', marker=dict(size=2,color = 'red')))

    fig.update_layout(
        xaxis_range=[limits[0], limits[1]],
        yaxis_range=[limits[2], limits[3]],
    )
    fig.show()

    return FINALPOINTS,WIDTHDATA

```
